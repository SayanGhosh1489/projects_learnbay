```python
#splitting the data
X = data.drop('claim_amount', axis = 1)
Y = data['claim_amount']
```

```python
#define column transformer for preprocessing
numerical_feature = X.select_dtypes(include = ['int64','float64']).columns
categorical_feature = X.select_dtypes(include = ['object']).columns
```

```python
print('Len of num feature: ',len(numerical_feature))
print('Len of cat feature: ',len(categorical_feature))
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
preprocessor = ColumnTransformer(transformers=[
    
    ('num', StandardScaler(), numerical_feature),
    ('cat', OneHotEncoder(), categorical_feature)
])
```

```python
#train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y, train_size= 0.8, random_state= 42)
```

```python
#correlation analysis
from scipy.stats import pearsonr
correlations = {}
for col in numerical_feature:
    corr,_ = pearsonr(X[col],Y)
    correlations[col] = corr
```

```python
dict(sorted(correlations.items(), key= lambda item: item[1], reverse= True))
```

```python
#chi-Seqare test
from scipy.stats import chi2_contingency
chi2_result = {}
for col in categorical_feature:
    contingency_table = pd.crosstab(X[col],Y)
    chi2,p,_,_ = chi2_contingency(contingency_table)
    chi2_result[col] = (chi2,p)
print("\nChi-Square Test:")
for key, value in chi2_result.items():
    print(f"{key} : chi2 = {value[0]:.4f}, P-value = {value[1]:.4f}")
```

```python
chi2_result
```

```python
#annova testing
from scipy.stats import f_oneway
annova_results = {}
for col in categorical_feature:
    categories = X[col].unique()
    groups = [Y[X[col] == category] for category in categories]
    f_val, p_val = f_oneway(*groups)
    annova_results[col] = (f_val, p_val)
    
print("\n ANOVA TEST:")
for key, value in annova_results.items():
    print(f"{key}: F = {value[0]: .4f}, p-value = {value[1]: .4f}")
```

## Model Building

```python
from sklearn.ensemble import AdaBoostRegressor
from sklearn.linear_model import LinearRegression, Ridge
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import StackingRegressor, VotingRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.utils import shuffle
```

```python
#difine models
from sklearn.pipeline import Pipeline
def get_model(regressor):
    model = Pipeline(steps=[
    
    ("preprocess", preprocessor),
    ("Regresoor", regressor)
])
    return model
```

```python
lr_model = get_model(LinearRegression())
ada_model = get_model(AdaBoostRegressor())
gra_model = get_model(GradientBoostingRegressor())
rf_model = get_model(RandomForestRegressor())
dt_model = get_model(DecisionTreeRegressor())
```

```python
estimator = [('lr', lr_model),
            ('rf', rf_model),
            ('gd', gra_model),
            ('ada', ada_model),
            ('dt', dt_model)]
stack_model = StackingRegressor(estimators=estimator,
                               final_estimator= Ridge())
vote_model = VotingRegressor(estimators=estimator)

#hyper parameter tunning
from sklearn.model_selection import GridSearchCV
```
